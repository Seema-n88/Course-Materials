```python
import numpy as np
import numpy.linalg as lng
import matplotlib.pyplot as plt
import numpy.random as rnd

X = np.array([[1, 1, 1], [-1, 0.3, 2]], dtype=float)
y = np.array([[-0.1, 0.5, 0.5]], dtype=float)
w = np.array([[-0.45], [0.2]], dtype=float)
all_weights = []
H = np.matmul(X, np.transpose(X))
b = np.matmul(-X, np.transpose(y))
learning_rate = 0.05
all_weights.append(w)
```

```python
print 'Constant learning rate'
for i in range(0, 3000):
    gradient = np.matmul(H, w) + b
    if lng.norm(gradient) <= 1.0E-6:
        print 'Algorithm converged in', i, 'steps'
        break
    w = w - learning_rate * gradient
    all_weights.append(w)

print 'Weights found with constant learning rate', w
w_0 = [all_weights[i][0, 0] for i in range(0, len(all_weights))]
w_1 = [all_weights[i][1, 0] for i in range(0, len(all_weights))]
x_axis = [i for i in range(0, len(all_weights))]

plt.figure()
plt.subplots_adjust(left=0.125, right=0.9, bottom=0.1, top=2.0, wspace=0.3, hspace=0.3)
plt.subplot(311)
plt.plot(w_0, w_1)
plt.xlabel('w0')
plt.ylabel('w1')
plt.subplot(312)
plt.plot(x_axis, w_0)
plt.xlabel('After iteration')
plt.ylabel('w0')
plt.subplot(313)
plt.plot(x_axis, w_1)
plt.xlabel('After iteration')
plt.ylabel('w1')
plt.show()
```

```python
all_weights = []
w = np.array([[-0.45], [0.2]], dtype=float)

all_weights.append(w)
print 'Line search'
for i in range(0, 3000):
    gradient = np.matmul(H, w) + b
    if lng.norm(gradient) <= 1.0E-6:
        print 'Algorithm converged in', i, 'steps'
        break
    learning_rate = np.matmul(np.transpose(gradient), gradient)[0, 0] / np.matmul(np.matmul(np.transpose(gradient), H), gradient)[0, 0]
    w = w - learning_rate * gradient
    all_weights.append(w)
print 'Weights found with line search method', w
w_0 = [all_weights[i][0, 0] for i in range(0, len(all_weights))]
w_1 = [all_weights[i][1, 0] for i in range(0, len(all_weights))]
x_axis = [i for i in range(0, len(all_weights))]


plt.figure()
plt.subplots_adjust(left=0.125, right=0.9, bottom=0.1, top=2.0, wspace=0.3, hspace=0.3)
plt.subplot(311)
plt.plot(w_0, w_1)
plt.xlabel('w0')
plt.ylabel('w1')
plt.subplot(312)
plt.plot(x_axis, w_0)
plt.xlabel('After iteration')
plt.ylabel('w0')
plt.subplot(313)
plt.plot(x_axis, w_1)
plt.xlabel('After iteration')
plt.ylabel('w1')
plt.show()
```

```python
print 'Conjugate gradient'
all_weights = []
w = np.array([[-0.45], [0.2]], dtype=float)
all_weights.append(w)
old_gradient = 0.0
gradient = np.matmul(H, w) + b
d = -gradient
for i in range(0, 3000):
    if lng.norm(gradient) <= 1.0E-6:
        print 'Algorithm converged in', i, 'steps'
        break
    step_size = - np.matmul(np.transpose(d), gradient)[0, 0] / np.matmul(np.matmul(np.transpose(d), H), d)
    w = w + step_size * d
    all_weights.append(w)
    old_gradient = gradient
    gradient = np.matmul(H, w) + b
    momentum = -np.matmul(np.transpose(gradient), gradient)[0, 0] / np.matmul(np.transpose(old_gradient), old_gradient)[0, 0]
    d = gradient + momentum * d
print 'Weights found with conjugate gradient', w
w_0 = [all_weights[i][0, 0] for i in range(0, len(all_weights))]
w_1 = [all_weights[i][1, 0] for i in range(0, len(all_weights))]
x_axis = [i for i in range(0, len(all_weights))]

plt.figure()
plt.subplots_adjust(left=0.125, right=0.9, bottom=0.1, top=2.0, wspace=0.3, hspace=0.3)
plt.subplot(311)
plt.plot(w_0, w_1)
plt.xlabel('w0')
plt.ylabel('w1')
plt.subplot(312)
plt.plot(x_axis, w_0)
plt.xlabel('After iteration')
plt.ylabel('w0')
plt.subplot(313)
plt.plot(x_axis, w_1)
plt.xlabel('After iteration')
plt.ylabel('w1')
plt.show()
```
