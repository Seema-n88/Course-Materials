## Exercise H7.2: k nearest neighbors (kNN)

```python
def get_k_nearest_neighbors(k, x_a, x):
    result = []
    index = 0
    for point in x_a:
        distance = np.linalg.norm(point-x)
        result.append([point[0], point[1], distance, index])
        index += 1
    result = np.array(result)
    return  result[result[:,2].argsort()][:k, :]
```

```python
def get_knn_class(k, train_set, x):
    index = get_k_nearest_neighbors(k, train_set[:,:2], x)[:,3]
    sum = 0
    for i in index:
        sum += train_set[int(i)][2:].T     
    result = sum/k
    return np.argmax(result)
```

```python
x_m = np.linspace(np.amin(x_1)-0.1,np.amax(x_1)+0.1,200)
y_m = np.linspace(np.amin(x_2)-0.1,np.amax(x_2)+0.1,200)
xx,yy = np.meshgrid(x_m, y_m, indexing="ij")
result_k1 = np.zeros([200,200])
result_k3 = np.zeros([200,200])
result_k5 = np.zeros([200,200])

for x_point,i in zip(x_m,range(len(x_m))):
    for y_point,j in zip(y_m,range(len(y_m))):
        knn_class1 = get_knn_class(1, train_set, [x_point,y_point])
        knn_class3 = get_knn_class(3, train_set, [x_point,y_point])
        knn_class5 = get_knn_class(5, train_set, [x_point,y_point])
        result_k1[i,j] = knn_class1
        result_k3[i,j] = knn_class3
        result_k5[i,j] = knn_class5
```

## Exercise H7.3:  ‚ÄúParzen window‚Äù classifier

```python
def get_parzen_window(x_a, x, variance):
    return np.exp(-0.5/variance*np.power(np.linalg.norm(x-x_a),2))

def get_parzen_class(train_set, x, variance):
    my_sum = 0
    z = 0
    for row in train_set:
        x_a = row[:2]
        y = row[2:].T
        p = get_parzen_window(x_a, x, variance)
        z += p
        my_sum += y*p
    return np.argmax(my_sum/z) # returns the index of the highest value -> most probable class
```

## Exercise H7.4: RBF networks

```python
import pandas as pd
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
p = 60 # number of samples from normal distributions for each label value y=0 and y=1

mu1, sigma = np.asarray([0,1]), np.sqrt(0.1) # mean1 and standard deviation
mu2, sigma = np.asarray([1,0]), np.sqrt(0.1) # mean2 and standard deviation

n1 = np.random.normal(mu1, sigma, size=[p, 2])
n2 = np.random.normal(mu2, sigma, size=[p, 2])

choice = np.random.choice([0,1],p)
sample1 = np.where(choice, n1.T, n2.T).T   # p(x|y = 0) = 0.5*[N(x|mu1,sigma) + N(x|mu2, sigma)]

mu3, sigma = np.asarray([0,0]), np.sqrt(0.1) # mean3 and standard deviation
mu4, sigma = np.asarray([1,1]), np.sqrt(0.1) # mean4 and standard deviation

n3 = np.random.normal(mu3, sigma, size=[p, 2])
n4 = np.random.normal(mu4, sigma, size=[p, 2])

choice = np.random.choice([0,1],p)
sample2 = np.where(choice, n3.T, n4.T).T   # p(x|y = 1) = 0.5*[N(x|mu3,sigma) + N(x|mu4, sigma)]

class0_data = np.concatenate((sample1.T, np.zeros((p, 1)).T), axis = 0).T
class1_data = np.concatenate((sample2.T, np.ones((p, 1)).T), axis = 0).T

plt.scatter(class0_data[:, 0], class0_data[:,1], label = "class0")
plt.scatter(class1_data[:, 0], class1_data[:,1], label = "class1",color = "red")
plt.legend()
```

```python
data = np.concatenate([class0_data, class1_data])
```

### Determine K- number of clusters using elbow method

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist

plt.rcParams['figure.figsize'] = (10, 6)
  
# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(data[:,:2])
    kmeanModel.fit(data[:,:2])
    distortions.append(sum(np.min(cdist(data[:,:2], kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / data[:,:2].shape[0])
 

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

As K increases, the data points are closer to the cluster's centroids.
The improvements regarding distance will decline at some point rapidly, creating the elbow shape.
That point is the optimal value for K. In the image above, K=4.

```python
def calculate_centroids(num_clusters,X):
    
    # Number of clusters
    kmeans = KMeans(n_clusters = num_clusters)
    # Fitting the input data
    kmeans = kmeans.fit(X)
    # Getting the cluster labels labels(0,1,2,3)
    labels = kmeans.predict(X)
    # Centroid values t0,...,t3
    centroids = kmeans.cluster_centers_
    
    return centroids, labels
```

```python
centroids,labels = calculate_centroids(4,data[:,:2])
centroids
```

```python
plt.style.use('ggplot')
plt.scatter(data[:, 0], data[:, 1], c = labels)
for i in range(centroids.shape[0]):
    plt.plot(centroids[i, 0], centroids[i,1], '*', color = 'red', ms = 8)
```

```python
def calculate_variances(k, centroids):
    std_devs = []
    
    for i in range(k):
        distances = np.array([np.linalg.norm(centroids[i] - centroids[j]) for j in range(k) if j != i])
        std_devs.append((1)*np.amin(distances))
        
    return np.square(std_devs)
```

```python
def calculate_phis(k, data, centroids, variances): 
    
    phis = np.zeros(shape = (k + 1, data.shape[0]))
    phis[0] = np.ones(shape = data.shape[0])
    
    for i in range(k):
        phis[i+1] = np.exp(-0.5 * ((np.linalg.norm(data[:,:2] - centroids[i], axis=1))) / variances[i])
        
    return phis
```

```python
def calculate_weights(phis, y):
    A = np.linalg.inv(np.dot(phis, phis.T))
    w = np.dot(np.dot(A, phis), y.T) 
    return w
```

```python
K = 4   # number of clusters
var = calculate_variances(K, centroids)
phis = calculate_phis(K, data, centroids, var)
Y = data[:,2]
w = np.array([calculate_weights(phis, Y)])
```

```python
n = 100
x = np.linspace(-1, 2, n)
X, Y = np.meshgrid(x, x)
points = np.vstack([X.flatten(), Y.flatten()]).T
```

```python
def prediction(k,variances):
    centroids,labels = calculate_centroids(k,data[:,:2])
    # print ("centroids for k = ", k, " are ", centroids)
    phis_x = calculate_phis(k, data[:,:2], centroids, variances)
    w = np.array([calculate_weights(phis_x, data[:,2])])
    phis_y = calculate_phis(k, points, centroids, variances)
    output = (np.dot(w, phis_y))
    # print("predictions min = ", output.min(), " and predictions max = ", output.max())
    y_hat = np.where(output >= 0.5, 1, 0)
    Z = y_hat.reshape(n, n)
    return Z, centroids

def plot_all(X, Y, Z, centorids):
    plt.contour(X, Y, Z)
    plt.scatter(class0_data[:, 0], class0_data[:,1], label = "class0", color = "lightblue")
    plt.scatter(class1_data[:, 0], class1_data[:,1], label = "class1",color = "lightgreen")
    for i in range(centroids.shape[0]):
        plt.plot(centroids[i, 0], centroids[i,1], '*', color = 'black', ms = 8)
    plt.legend()
    
```

```python
plt.figure(figsize=(16,6))
plt.style.use('ggplot')

plt.subplot(1,3,1)
Z, centroids = prediction(2, [0.5, 0.5])
plot_all(X, Y, Z, centroids)
plt.title('Number of clusters is 2')

plt.subplot(1,3,2)
Z, centroids = prediction(3, [0.5, 0.5, 0.5])
plot_all(X, Y, Z, centroids)
plt.title('Number of clusters is 3')

plt.subplot(1,3,3)
Z, centroids = prediction(4, [0.5, 0.5, 0.5, 0.5])
plot_all(X, Y, Z, centroids)
plt.title('Number of clusters is 4')
```
