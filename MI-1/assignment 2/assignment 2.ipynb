```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
from sklearn.metrics import accuracy_score
import math
%matplotlib inline

data = np.genfromtxt('applesOranges.csv',delimiter=',', skip_header=1)

plt.figure(figsize = (10,10))
ax = plt.subplot(111)
color = ['red' if l == 0 else'blue' for l in data[:,2]]
plt.scatter(data[:,0], data[:,1], c =color)

y = [1 if l > 0 else -1 for l in data[:,2]]
print y
```

```python
theta = 0

phi = np.linspace(0,np.pi,19)
w1,w2 = np.cos(phi),np.sin(phi)
w = np.vstack((w1,w2))
plt.figure(figsize = (10,5))
plt.plot(w1,w2,'.')
plt.plot((0,w1[2]),(0,w2[2]))
```

```python
numSamples = len(data)

def getaccuracy(yhat,y):
    print yhat
    print y
    tmp = np.sum(np.abs(yhat-y))
    print tmp
    return tmp/numSamples

x1,x2 = data[:,0],data[:,1]
x = np.vstack((x1,x2))
z = np.dot(w.T,x) - theta
y_w = np.sign(z)

perf = [accuracy_score(y, yw) for yw in y_w]

print perf
m = np.argmax(perf)

alpha = [np.round(math.degrees(phi[x]),2) for x in range(len(phi))]
plt.figure(figsize = (10,7))
plt.plot(alpha, perf)
plt.plot((alpha[m],alpha[m]), (min(perf),max(perf)+0.1), 'r')

```
